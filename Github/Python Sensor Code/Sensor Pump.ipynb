```python
import paho.mqtt.client as mqtt
import time
import random

BROKER = "broker.hivemq.com"  # ganti ke broker lokal kalau mau
PORT = 1883

client = mqtt.Client(client_id="pump-sensor-01", protocol=mqtt.MQTTv311, transport="tcp", userdata=None, clean_session=True)

client.connect(BROKER, PORT, 60)

while True:
    pressure = round(random.uniform(2.5, 3.5), 2)
    flow = round(random.uniform(100, 150), 1)
    status = random.choice(["ON", "OFF"])

    client.publish("pump/1/pressure", pressure)
    client.publish("pump/1/flow", flow)
    client.publish("pump/1/status", status)

    print(f"Published: Pressure={pressure} bar | Flow={flow} L/min | Status={status}")
    time.sleep(5)  # kirim tiap 5 detik
```
